```python
import requests
```

```python
def get_html_data(url):
    response = requests.get(url)

    if response.status_code == 200:
        html_data = response.text
        return html_data
    else:
        print("error")
        return None
```

```python
url = "https://www.westmetall.com/en/markdaten.php"
```

```python
html_data = get_html_data(url)
```

```python
if html_data:
    print(html_data)
```
