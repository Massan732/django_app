```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt 
import japanize_matplotlib
```

```python
df = pd.read_csv("テストデータ.csv", encoding="cp932")
```

```python
df['金額'] = df['金額'].fillna(0).str.replace("nan","0")
df['金額'] = df["金額"].str.replace(',', '').astype(float)
```

```python
df['重量'] = df['重量'].fillna(0).str.replace("nan","0")
df['重量'] = df["重量"].str.replace(',', '').astype(float)
```

```python
date_filter = df.groupby(["受注日"]).agg({"金額":"sum"}).reset_index()
```

```python
#date_filter = date_filter.sort_values('金額', ascending=False)
```

```python
date_filter
```

```python
# 二次曲線のプロット作成
plt.plot(date_filter["受注日"], date_filter["金額"], label="棒")
 
# タイトル・軸ラベル表示
plt.title("グラフタイトル")
plt.xlabel("x軸ラベル名")
plt.ylabel("y軸ラベル名")
plt.xticks(rotation=90)
# グラフ内テキスト表示
plt.text(0, 4,"テキスト例")
 
# 凡例表示
plt.legend()
 
plt.show()
```

```python
date_filter = df.groupby(["担当者名"]).agg({"金額":"sum"}).reset_index()
```

```python
# 二次曲線のプロット作成
plt.bar(date_filter["担当者名"], date_filter["金額"], label="棒")
 
# タイトル・軸ラベル表示
plt.title("グラフタイトル")
plt.xlabel("x軸ラベル名")
plt.ylabel("y軸ラベル名")
plt.xticks(rotation=90)
# グラフ内テキスト表示
plt.text(0, 4,"テキスト例")
 
# 凡例表示
plt.legend()
 
plt.show()
```
