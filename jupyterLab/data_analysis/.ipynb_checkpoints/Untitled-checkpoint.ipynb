```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv("テストデータ.csv", encoding="cp932")
```

```python
df['金額'] = df['金額'].fillna(0).str.replace("nan","0")
```

```python
df['金額'] = df["金額"].str.replace(",","").astype(int)
```

```python
date_filter = df.groupby(["取引先"]).agg({"金額":"sum"}).reset_index()
```

```python
date_filter = date_filter.sort_values(by='金額')
```

```python
import matplotlib as mpl
print(mpl.matplotlib_fname())
```

```python
# 折れ線グラフのトレースを作成
trace = go.Scatter(x=date_filter['取引先'], y=date_filter['金額'], mode='lines+markers', name='金額')

# レイアウトを設定
layout = go.Layout(title='売上データの折れ線グラフ', xaxis=dict(title='年度'), yaxis=dict(title='売上額'))

# グラフオブジェクトを作成
fig = go.Figure(data=[trace], layout=layout)
```

```python
fig.show()
```
